## Splitting the data for testing and training

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.8,random_state=10)
```

```python
len(X_train),len(X_test)
```

## Model Building

### Using GridSearchCV to find the best possible models for our data set

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier

model_params = {
    'svm': {
        'model': svm.SVC(gamma='auto'),
        'params' : {
            'C': [1,10,20],
            'kernel': ['rbf','linear','poly']
        }
    },  

    'decision_trees': {
        'model': DecisionTreeClassifier(),
        'params' : {
            'criterion': ['gini','entropy'],
            'splitter' : ['best','random'] 

            
        }
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'n_estimators': [1,5,10,30,40]
        }
    },
    'logistic_regression' : {
        'model': LogisticRegression(solver='liblinear',multi_class='auto'),
        'params': {
            'C': [1,5,10]
        }
    },
    'Gaussian' : {
        'model' : GaussianNB(),
        'params': {}
    },

    'multi_Nomial' : {
        'model' : MultinomialNB(),
        'params': {}    
        }
    
}
```

```python
from sklearn.model_selection import GridSearchCV
scores = []

for model_name, mp in model_params.items():
    clf =  GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    clf.fit(X_train, y_train)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
    
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```

## Using the GridSearchCV it has been found that Random Forest is best for our dataset so we select it for our model

# Random Forest

```python
rf = RandomForestClassifier(n_estimators=40,random_state=10)
rf.fit(X_train,y_train)
rf.score(X_test,y_test)
```

## The accuracy of Random Forest Algorith is Highest i.e (99%), Hence we select it

### Using K-Fold to confirm the accuracy

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(RandomForestClassifier(n_estimators=40),X,y,cv=5)
```

### Using confusion matrix for error evaluation

```python
from sklearn.metrics import confusion_matrix
y_predicted = rf.predict(X_test)
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sn

plt.figure(figsize = (3,3))

sn.heatmap(cm, annot=True)

plt.xlabel('Predicted')
plt.ylabel('Truth')



```

### Using the Heatmap, out of 104 Test cases our model predicted 103 cases correctly and failed only once, hence we conclude with Random Forest Algorithm

## Feature Score calculation

```python

# get importance
importance = rf.feature_importances_
ypos = np.arange(len(importance))
plt.yticks(ypos,X.columns,fontsize=15)
plt.rcParams['figure.figsize']=(20,12)
# summarize feature importance
for i,v in enumerate(importance):
    print('Feature: %0d, Score: %.5f' % (i,v))
# plot feature importance
plt.xlabel('Score',fontsize=18)
plt.ylabel('Features',fontsize=18)
plt.title('Feature Scores',fontsize=22)

plt.barh(ypos, importance)
plt.show()
```
